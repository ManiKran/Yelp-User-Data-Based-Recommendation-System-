# Data Preparation and Cleaning

In the start, we have three different Datasets. now we are trying to merge the datasets by renaming the columns names where two datasets
have the same name.

```python
import pandas as pd
import numpy as np

#tip_df = pd.read_csv("/Users/student/Downloads/yelp_dataset/yelp_academic_dataset_tip.csv")
#checkin_df = pd.read_csv("/Users/student/Downloads/yelp_dataset/yelp_academic_dataset_checkin.csv")
print("done")
user_df = pd.read_csv("/Users/student/Downloads/yelp_dataset/yelp_academic_dataset_user.csv", low_memory = False)
print("done")
review_df = pd.read_csv("/Users/student/Downloads/yelp_dataset/yelp_academic_dataset_review.csv")
print("done")
business_df = pd.read_csv("/Users/student/Downloads/yelp_dataset/yelp_academic_dataset_business.csv")
print("done")
```

```python
common_columns_in_user_review = set(user_df.columns.tolist()) & set(review_df.columns.tolist())
print(common_columns_in_user_review)
```

```python
columns_to_print = ['useful', 'funny', 'cool', 'user_id']
user_df[columns_to_print].head()
```

```python
review_df[columns_to_print].head()
```

```python
columns_to_print = ['no_of_useful_reviews_by_user', 'no_of_funny_reviews_by_user', 'no_of_cool_reviews_by_user', 'user_id']
user_df.rename(columns={'useful': 'no_of_useful_reviews_by_user'}, inplace=True)
user_df.rename(columns={'funny': 'no_of_funny_reviews_by_user'}, inplace=True)
user_df.rename(columns={'cool': 'no_of_cool_reviews_by_user'}, inplace=True)
user_df[columns_to_print].head()
```

```python
columns_to_print = ['is_useful', 'is_funny', 'is_cool', 'user_id']
review_df.rename(columns={'useful': 'is_useful'}, inplace=True)
review_df.rename(columns={'funny': 'is_funny'}, inplace=True)
review_df.rename(columns={'cool': 'is_cool'}, inplace=True)
review_df[columns_to_print].head()
```

```python
user_review_merged_df = pd.merge(user_df, review_df, on='user_id')
```

```python
user_review_merged_df.shape
```

```python
common_columns_in_user_review_business = set(user_review_merged_df.columns.tolist()) & set(business_df.columns.tolist())
print(common_columns_in_user_review_business)
```

```python
columns_to_print = ['review_count', 'stars', 'name', 'business_id']
user_review_merged_df[columns_to_print].head()
```

```python
columns_to_print = ['review_count', 'stars', 'name', 'business_id']
business_df[columns_to_print].head()
```

```python
columns_to_print = ['no_of_reviews_by_user', 'rating_by_user', 'user_name', 'business_id']
user_review_merged_df.rename(columns={'review_count': 'no_of_reviews_by_user'}, inplace=True)
user_review_merged_df.rename(columns={'stars': 'rating_by_user'}, inplace=True)
user_review_merged_df.rename(columns={'name': 'user_name'}, inplace=True)
user_review_merged_df[columns_to_print].head()
```

```python
columns_to_print = ['no_of_reviews_for_business', 'rating_for_business', 'business_name', 'business_id']
business_df.rename(columns={'review_count': 'no_of_reviews_for_business'}, inplace=True)
business_df.rename(columns={'stars': 'rating_for_business'}, inplace=True)
business_df.rename(columns={'name': 'business_name'}, inplace=True)
business_df[columns_to_print].head()
```

```python
complete_merged_df = pd.merge(business_df, user_review_merged_df, on = 'business_id')
```

```python
complete_merged_df.shape
```

```python
complete_merged_df.head()
```

```python
final_df = complete_merged_df.copy()
```

Additionally, we are focusing solely on businesses located in the city of "Santa Barbara". After filtering, we will save the resulting DataFrame.

```python
final_df = final_df[final_df['categories'].str.contains('Restaurants|Food', case=False)]
```
